```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
```

```python
test_df=pd.read_csv('test.csv')
train_df=pd.read_csv('train.csv')
```

```python
# ONE-HOT ENCODING
train_df=pd.get_dummies(train_df,columns=['Sex'])
test_df=pd.get_dummies(test_df,columns=['Sex'])
train_df=pd.get_dummies(train_df,columns=['Embarked'])
test_df=pd.get_dummies(test_df,columns=['Embarked'])
# FILL MISSING AGE WITH MEDIAN
train_df['Age']=train_df['Age'].fillna(train_df['Age'].median())
test_df['Age']=test_df['Age'].fillna(test_df['Age'].median())
# FILL FARE WITH MEDIAN IN TEST DATA
test_df['Fare']=test_df['Fare'].fillna(test_df['Fare'].median())
# MIN-MAX-SCALER OF FARE in train_df
scaler=MinMaxScaler()
train_df['Fare']=scaler.fit_transform(train_df[['Fare']])
train_df['Pclass']=scaler.fit_transform(train_df[['Pclass']])
train_df['Age']=scaler.fit_transform(train_df[['Age']])
train_df['SibSp']=scaler.fit_transform(train_df[['Fare']])
train_df['Parch']=scaler.fit_transform(train_df[['Parch']])
# MIN-MAX-SCALER OF FARE in test_df
test_df['Fare']=scaler.fit_transform(test_df[['Fare']])
test_df['Pclass']=scaler.fit_transform(test_df[['Pclass']])
test_df['Age']=scaler.fit_transform(test_df[['Age']])
test_df['SibSp']=scaler.fit_transform(test_df[['Fare']])
test_df['Parch']=scaler.fit_transform(test_df[['Parch']])
```
